```python
X = bodyfat.iloc[:, 5:]
```

```python
y = bodyfat.iloc[:, 1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, RobustScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import ExtraTreesRegressor, GradientBoostingRegressor
from sklearn import metrics
```

```python
def preprocess(X, y, rand):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = rand)
    mct = make_column_transformer((StandardScaler(), X.columns))
    X_train = mct.fit_transform(X_train)
    X_test = mct.transform(X_test)
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocess(X, y, 101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.score(X_test, y_test)
```

```python
pred = lm.predict(X_test)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, pred))
lm.coef_
```

```python
sns.regplot(x=y_test, y=pred)
```

```python
import sklearn
from sklearn.utils import all_estimators
```

```python
regressors = []
for name, class_ in all_estimators():
    if issubclass(class_, sklearn.base.RegressorMixin):
        regressors.append((name, class_))
```

```python
len(regressors)
```

```python
scores = {}
errors = {}
# ARDRegression, BayesianRidge, GaussianProcessRegressor and KernelRidge are taking too much memory.
for i, j in regressors[1:3] + regressors[4:11] + regressors[12:17] + regressors[18:]:
    if i == 'DummyRegressor':
        continue
    try:
        X_train, X_test, y_train, y_test = preprocess(X, y, 101)
        reg = j()
        if hasattr(reg, 'copy_X'):
            reg.copy_X = False
        reg.fit(X_train, y_train)
        scores[i] = reg.score(X_test, y_test)
    except Exception as e:
        errors[j] = e
```

```python
scores = pd.DataFrame(data={'estimator':scores.keys(), 'score':scores.values()})
```

```python
scores.sort_values('score', ascending=False, inplace=True)
```

```python
scores
```

```python
plt.figure(figsize=(20,6))
plt.plot(scores['estimator'], scores['score'])
plt.xticks(rotation=65)
for i, v in enumerate(scores['score']):
    plt.text(i-.25, v+.01, np.round(v,3))
plt.grid()
plt.show()
```
